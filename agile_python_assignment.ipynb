```python
import pandas as pd
import numpy as np
```

```python
input1=pd.read_csv('input1.csv')
```

```python
input2=pd.read_csv('input2.csv')
```

```python
input1
```

```python
input2
```

```python
new=input1.merge(input2,left_on='User ID',right_on='uid')
```

```python
out1=new.groupby('Team Name')
```

```python
out1
```

```python
mod=out1['total_statements','total_reasons'].mean()
```

```python
mod.sort_values(['total_statements','total_reasons'],ascending=[False,False])
```

```python
output1=mod.rename(columns={'Team Name':'Thinking Teams Leaderboard','total_statements':'Average Statements','total_reasons':'Average Reasons'})
```

```python
output1
```

```python
output1.rename(index={'Team Name':'Thinking Teams Leaderboard'})
```

```python
output1=output1.sort_values(['Average Statements','Average Reasons'],ascending=[False,False])
```

```python
start=1
output1.insert(0,'RANK',range(start,start+output1.shape[0]))
output1
```

```python
output2=input2.sort_values(['total_statements','total_reasons'],ascending=[False,False])
```

```python
start=1
output2.insert(0,'Rank',range(start,start+output2.shape[0]))
output2
```

```python
output2.drop(['S No'],axis=1,inplace=True)
```

```python
output2.rename(columns={'total_statements':'No of Statements','total_reasons':'No of Reasons'},inplace=True)
```
